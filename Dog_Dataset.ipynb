```python
import pandas as pd

df = pd.read_csv("dogs_dataset.csv")

df.head
```

```python
breed_avg_weight = df.groupby("Breed")["Weight (kg)"].mean().sort_values(ascending=False)

#10 Razas más pesadas
print("Top 10 Heaviest Breeds")
print(breed_avg_weight.head(10))
```

```python
#Visualización con Matplot
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
breed_avg_weight.head(10).plot(kind="barh", color="darkorange")
plt.xlabel("Average Weight (kg)")
plt.title("Top 10 Heaviest Dog Breeds")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
#Boxplot con seaborn
import seaborn as sns
import matplotlib.pyplot as plt

top_breeds = breed_avg_weight.head(10).index
filtered_df = df[df["Breed"].isin(top_breeds)]

plt.figure(figsize=(12, 6 ))
sns.boxplot(data=filtered_df, x="Breed", y="Weight (kg)", palette="pastel")
plt.xticks(rotation=45)
plt.title("Weight Distribution of Top 10 Heaviest Breeds")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("dogs_dataset.csv")
breed_avg_weight = df.groupby("Breed")["Weight (kg)"].mean().sort_values()
```

```python
lightest = breed_avg_weight.head(10)
heaviest = breed_avg_weight.tail(10)
combined = pd.concat([lightest, heaviest])
combined_df = combined.reset_index()
combined_df["Category"] = ["Lightest"] *10 + ["Heaviest"] *10
```

```python
#Bar Chart comparación

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,6))
sns.barplot(data=combined_df, x="Weight (kg)", y="Breed", hue="Category", palette="coolwarm")
plt.title("Top 10 Lightests vs Heaviest Dogs")
plt.xlabel("Average Weight (kg)")
plt.ylabel ("Breed")
plt.tight_layout()
plt.show()
```
